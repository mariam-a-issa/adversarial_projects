```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = np.array([
    [0.8626, 0.8809, 0.027, 0],
    [0.9286, 0.9329, 0.88, 0.859],
    [0.901714,	0.904229,	0.853314,	0.8570],
    [0.962457,	0.95400,	0.862686,	0.834],
    [0.993600,	0.99381,	0.871300,	0.827]
])

# data = np.array(((86.26, 88.09, 0.27, 0), 
#                  (92.86, 93.29, 88, 85.9)))
# data
```

```python
df = pd.DataFrame(data, 
                  index=['CNN', 'HD', 'PGD', 'CAE', "FHD"], 
                  columns=['FGSM', 'DF', 'JSMA', 'GA'])
```

```python
df
```

```python
plt.figure(figsize=(15, 8))

plt.plot(df.iloc[0, :], label='CNN')
plt.plot(df.iloc[1, :], label='HD')
plt.plot(df.iloc[2, :], label='PGD')
plt.plot(df.iloc[3, :], label='CAE')
plt.plot(df.iloc[4, :], label='FHD')

plt.ylim(40, 100)
plt.title("Adversarial Sample Accuracy")
plt.grid()
plt.legend()
```
